```python
from torch.utils.data import Dataset
from PIL import Image
from torchvision import transforms
import glob
import os

class SRDataset(Dataset):
    def __init__(self, hr_folder, scale=4):
        self.hr_paths = sorted(glob.glob(os.path.join(hr_folder, "*.png")))
        self.scale = scale
        self.to_tensor = transforms.ToTensor()

    def __len__(self):
        return len(self.hr_paths)

    def __getitem__(self, idx):
        hr = Image.open(self.hr_paths[idx]).convert("RGB")

        # Generate LR image using bicubic downsampling
        lr = hr.resize(
            (hr.width // self.scale, hr.height // self.scale),
            Image.BICUBIC
        )

        return self.to_tensor(lr), self.to_tensor(hr)
```

```python
from torch.utils.data import Dataset, DataLoader
from PIL import Image
from torchvision import transforms
import glob
import os

class SRDataset(Dataset):
    def __init__(self, hr_folder, scale=4):
        self.hr_paths = sorted(glob.glob(os.path.join(hr_folder, "*.png")))
        self.scale = scale

        self.hr_transform = transforms.Compose([
            transforms.Resize((256, 256)),   # Make every HR image 256×256
            transforms.ToTensor()
        ])

    def __len__(self):
        return len(self.hr_paths)

    def __getitem__(self, idx):
        hr = Image.open(self.hr_paths[idx]).convert("RGB")
        hr = self.hr_transform(hr)

        lr = transforms.functional.resize(
            hr,
            [64, 64],
            interpolation=transforms.InterpolationMode.BICUBIC
        )

        return lr, hr
```

```python
hr_folder = "/kaggle/input/datasets/soumikrakshit/div2k-high-resolution-images/DIV2K_train_HR/DIV2K_train_HR"

train_dataset = SRDataset(hr_folder)

train_loader = DataLoader(
    train_dataset,
    batch_size=4,
    shuffle=True,
    num_workers=0
)
```

```python
lr_img, hr_img = next(iter(train_loader))

print(lr_img.shape)
print(hr_img.shape)
```

```python
from models.network_swinir import SwinIR
```

```python


model = SwinIR(
    upscale=4,
    in_chans=3,
    img_size=64,
    window_size=8,
    img_range=1.0,
    depths=[6,6,6,6],
    embed_dim=30,
    num_heads=[3,3,3,3],  # 30 divisible by 3
    mlp_ratio=2,
    upsampler='pixelshuffle',
    resi_connection='1conv'
)
```

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = model.to(device)

print(device)
```

```python
lr_img = lr_img.to(device)

with torch.no_grad():
    sr_img = model(lr_img)

print(sr_img.shape)
```

```python
print(sum(p.numel() for p in model.parameters() if p.requires_grad) / 1e6)
```

```python
import torch.nn as nn

optimizer = torch.optim.Adam(model.parameters(), lr=2e-4)
loss_fn = nn.L1Loss()

num_epochs = 20    # <-- Define number of epochs here

for epoch in range(num_epochs):

    model.train()
    epoch_loss = 0

    for lr_img, hr_img in train_loader:

        lr_img = lr_img.to(device)
        hr_img = hr_img.to(device)

        sr_img = model(lr_img)

        loss = loss_fn(sr_img, hr_img)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

    print(f"Epoch {epoch+1}/{num_epochs} Loss: {epoch_loss/len(train_loader):.4f}")

    torch.save(
        model.state_dict(),
        f"/kaggle/working/swinir_epoch_{epoch+1}.pth"
    )
```

```python
from PIL import Image
from torchvision import transforms
from skimage.metrics import peak_signal_noise_ratio, structural_similarity
import numpy as np
import torch

# Validation image
img_path = "/kaggle/input/datasets/soumikrakshit/div2k-high-resolution-images/DIV2K_valid_HR/DIV2K_valid_HR/0801.png"

# Load HR image
hr = Image.open(img_path).convert("RGB")
hr = hr.resize((256, 256))

# Create LR image
lr = hr.resize((64, 64), Image.BICUBIC)

# Convert LR to tensor
to_tensor = transforms.ToTensor()
lr_tensor = to_tensor(lr).unsqueeze(0).to(device)

# Inference
model.eval()
with torch.no_grad():
    sr = model(lr_tensor)

# Convert SR to numpy
sr = sr.squeeze().cpu().permute(1, 2, 0).numpy()
sr = np.clip(sr, 0, 1)

# Convert HR to numpy
hr_np = np.array(hr).astype(np.float32) / 255.0

# PSNR
psnr = peak_signal_noise_ratio(hr_np, sr, data_range=1.0)

# SSIM
ssim = structural_similarity(
    hr_np,
    sr,
    channel_axis=2,
    data_range=1.0
)

print("PSNR =", psnr)
print("SSIM =", ssim)
```
